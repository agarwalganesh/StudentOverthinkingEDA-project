## 1️⃣ Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# Set style for better visualizations
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

print("✅ All libraries imported successfully!")
```

## 2️⃣ Load & Display Basic Data Information

```python
# Load dataset
import os
os.chdir(r'c:\Users\ganes\OneDrive\Pictures\Documents\Desktop\MAC_PRO Project\mental health')
df = pd.read_csv('Student Mental health.csv')

print("="*70)
print("📋 DATASET OVERVIEW".center(70))
print("="*70)

print(f"\n📊 Shape: {df.shape[0]} rows × {df.shape[1]} columns")
print(f"\n📌 Column Names & Types:")
print(df.dtypes)

print(f"\n\n📖 First 5 Rows:")
print(df.head())

print(f"\n\n🔍 Dataset Info:")
print(df.info())
```

## 3️⃣ Missing Values Analysis

```python
print("="*70)
print("🔎 MISSING VALUES ANALYSIS".center(70))
print("="*70)


# Calculate missing values
missing_data = pd.DataFrame({
    'Column': df.columns,
    'Missing_Count': df.isnull().sum().values,
    'Missing_Percentage': (df.isnull().sum().values / len(df) * 100).round(2)
})

print(f"\n{missing_data.to_string(index=False)}")

total_missing = df.isnull().sum().sum()
print(f"\n\n📊 Total Missing Values: {total_missing}")

if total_missing == 0:
    print("✅ No missing values found!")
else:
    print(f"⚠️  {total_missing} missing values need attention")
```

## 4️⃣ Statistical Summary

```python
print("="*70)
print("📈 STATISTICAL SUMMARY".center(70))
print("="*70)

print(f"\n🔢 Numeric Columns Summary:")
# print(df.to_string(index=False))
numeric_df = df.select_dtypes(include=[np.number])
numeric_stats = numeric_df.describe().T
print(numeric_stats)

# Additional statistics
print(f"\n\n📊 Additional Statistics:")
print(f"\nSkewness:")
print(numeric_df.skew())

print(f"\n\nKurtosis:")
print(numeric_df.kurtosis())
```

## 5️⃣ Categorical Variables Analysis

```python
print("="*70)
print("🏷️  CATEGORICAL VARIABLES ANALYSIS".center(70))
print("="*70)

categorical_cols = df.select_dtypes(include=['object']).columns

for col in categorical_cols:
    print(f"\n\n📌 {col}:")
    print(f"   Unique Values: {df[col].nunique()}")
    print(f"\n   Value Counts:")
    value_counts = df[col].value_counts()
    for idx, (value, count) in enumerate(value_counts.items(), 1):
        percentage = (count / len(df)) * 100
        print(f" {idx}. {value}: {count} ({percentage:.1f}%)")
```

## 6️⃣ Distribution Analysis - Histograms

```python
print("\n📊 Creating Distribution Plots...\n")

numeric_cols = df.select_dtypes(include=[np.number]).columns
num_cols = len(numeric_cols)

fig, axes = plt.subplots(2, 3, figsize=(15, 10))
fig.suptitle('Distribution Analysis - Histograms', fontsize=16, fontweight='bold', y=1.00)

axes = axes.flatten()

for idx, col in enumerate(numeric_cols):
    ax = axes[idx]
    
    # Create histogram with KDE
    data = df[col].dropna()
    ax.hist(data, bins=20, alpha=0.7, color='steelblue', edgecolor='black')
    ax.set_title(f'{col}', fontweight='bold', fontsize=11)
    ax.set_xlabel('Value')
    ax.set_ylabel('Frequency')
    ax.grid(alpha=0.3)
    
    # Add statistics
    mean_val = data.mean()
    std_val = data.std()
    ax.axvline(mean_val, color='red', linestyle='--', linewidth=2, label=f'Mean: {mean_val:.2f}')
    ax.legend(loc='upper right')

# Remove extra subplots
for idx in range(num_cols, len(axes)):
    fig.delaxes(axes[idx])

plt.tight_layout()
plt.savefig('EDA_Distributions.png', dpi=300, bbox_inches='tight')
plt.show()

print("✅ Distribution plots saved as 'EDA_Distributions.png'")
```

## 7️⃣ Box Plots - Outliers Detection

```python
print("\n📦 Creating Box Plots for Outlier Detection...\n")

numeric_cols = df.select_dtypes(include=[np.number]).columns
fig, axes = plt.subplots(1, len(numeric_cols), figsize=(15, 5))
fig.suptitle('Box Plots - Outlier Detection', fontsize=16, fontweight='bold')

# Handle single column case - convert to array
if len(numeric_cols) == 1:
    axes = [axes]

for idx, col in enumerate(numeric_cols):
    ax = axes[idx]
    data = df[col].dropna()
    
    bp = ax.boxplot(data, vert=True, patch_artist=True)
    bp['boxes'][0].set_facecolor('lightblue')
    
    ax.set_title(f'{col}', fontweight='bold')
    ax.set_ylabel('Value')
    ax.grid(alpha=0.3, axis='y')

plt.tight_layout()
plt.savefig('EDA_BoxPlots.png', dpi=300, bbox_inches='tight')
plt.show()

print("✅ Box plots saved as 'EDA_BoxPlots.png'")
```

## 8️⃣ Categorical Variables - Count Plots

```python
print("\n🏷️  Creating Count Plots for Categorical Variables...\n")

categorical_cols = df.select_dtypes(include=['object']).columns
num_cat = len(categorical_cols)

# Calculate grid size dynamically based on number of categories
import math
num_cols = 4
num_rows = math.ceil(num_cat / num_cols)

fig, axes = plt.subplots(num_rows, num_cols, figsize=(16, 5*num_rows))
fig.suptitle('Categorical Variables - Distribution', fontsize=16, fontweight='bold', y=0.995)

axes = axes.flatten()

for idx, col in enumerate(categorical_cols):
    ax = axes[idx]
    
    value_counts = df[col].value_counts()
    colors = plt.cm.Set3(np.linspace(0, 1, len(value_counts)))
    
    value_counts.plot(kind='bar', ax=ax, color=colors, edgecolor='black')
    ax.set_title(f'{col}', fontweight='bold', fontsize=10)
    ax.set_xlabel('')
    ax.set_ylabel('Count')
    ax.tick_params(axis='x', rotation=45)
    ax.grid(alpha=0.3, axis='y')

# Remove extra subplots
for idx in range(num_cat, len(axes)):
    fig.delaxes(axes[idx])

plt.tight_layout()
plt.savefig('EDA_Categorical.png', dpi=300, bbox_inches='tight')
plt.show()

print("✅ Categorical plots saved as 'EDA_Categorical.png'")
```

## 9️⃣ Correlation Analysis

```python
print("="*70)
print("🔗 CORRELATION ANALYSIS".center(70))
print("="*70)

# Select only numeric columns
numeric_df = df.select_dtypes(include=[np.number])

print(f"\n📊 Correlation Matrix:")
correlation_matrix = numeric_df.corr()
print(correlation_matrix)

# Find high correlations
print(f"\n\n🔍 High Correlations (|r| > 0.5):")
high_corr = []
for i in range(len(correlation_matrix.columns)):
    for j in range(i+1, len(correlation_matrix.columns)):
        corr_val = correlation_matrix.iloc[i, j]
        if abs(corr_val) > 0.5:
            high_corr.append({
                'Variable 1': correlation_matrix.columns[i],
                'Variable 2': correlation_matrix.columns[j],
                'Correlation': corr_val
            })

if high_corr:
    high_corr_df = pd.DataFrame(high_corr)
    print(high_corr_df.to_string(index=False))
else:
    print("   No high correlations found.")
```

## 🔟 Correlation Heatmap

```python
print("\n🔥 Creating Correlation Heatmap...\n")

fig, ax = plt.subplots(figsize=(8, 6))

# Create heatmap
correlation_matrix = numeric_df.corr()
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', 
            center=0, square=True, linewidths=1, cbar_kws={"shrink": 0.8},
            ax=ax)

ax.set_title('Correlation Heatmap - Numeric Variables', fontsize=14, fontweight='bold', pad=20)
plt.tight_layout()
plt.savefig('EDA_Heatmap.png', dpi=300, bbox_inches='tight')
plt.show()

print("✅ Heatmap saved as 'EDA_Heatmap.png'")
```

## 1️⃣1️⃣ Age Distribution Analysis

```python
print("\n" + "="*70)
print("📍 AGE DISTRIBUTION ANALYSIS".center(70))
print("="*70)

age_data = df['Age'].dropna()

print(f"\n📊 Age Statistics:")
print(f"   Mean Age: {age_data.mean():.2f}")
print(f"   Median Age: {age_data.median():.2f}")
print(f"   Std Dev: {age_data.std():.2f}")
print(f"   Min Age: {age_data.min():.2f}")
print(f"   Max Age: {age_data.max():.2f}")
print(f"   Q1 (25%): {age_data.quantile(0.25):.2f}")
print(f"   Q3 (75%): {age_data.quantile(0.75):.2f}")
print(f"   IQR: {age_data.quantile(0.75) - age_data.quantile(0.25):.2f}")

# Age groups
print(f"\n\n👥 Age Groups:")
age_groups = pd.cut(age_data, bins=[0, 18, 20, 22, 25, 100], 
                     labels=['<18', '18-20', '20-22', '22-25', '>25'])
print(age_groups.value_counts().sort_index())

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Histogram with KDE
axes[0].hist(age_data, bins=15, alpha=0.7, color='skyblue', edgecolor='black')
axes[0].axvline(age_data.mean(), color='red', linestyle='--', linewidth=2, label=f'Mean: {age_data.mean():.2f}')
axes[0].axvline(age_data.median(), color='green', linestyle='--', linewidth=2, label=f'Median: {age_data.median():.2f}')
axes[0].set_title('Age Distribution', fontweight='bold', fontsize=12)
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Frequency')
axes[0].legend()
axes[0].grid(alpha=0.3)

# Age groups bar chart
age_groups.value_counts().sort_index().plot(kind='bar', ax=axes[1], color='coral', edgecolor='black')
axes[1].set_title('Age Groups Distribution', fontweight='bold', fontsize=12)
axes[1].set_xlabel('Age Group')
axes[1].set_ylabel('Count')
axes[1].tick_params(axis='x', rotation=45)
axes[1].grid(alpha=0.3, axis='y')

plt.tight_layout()
plt.savefig('EDA_Age_Analysis.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n✅ Age analysis plot saved as 'EDA_Age_Analysis.png'")
```

## 1️⃣2️⃣ Mental Health Conditions - Cross Analysis

```python
print("\n" + "="*70)
print("🧠 MENTAL HEALTH CONDITIONS ANALYSIS".center(70))
print("="*70)

mental_health_cols = ['Do you have Depression?', 'Do you have Anxiety?', 'Do you have Panic attack?']

print(f"\n📊 Individual Condition Counts:")
for col in mental_health_cols:
    yes_count = (df[col] == 'Yes').sum()
    no_count = (df[col] == 'No').sum()
    yes_pct = (yes_count / len(df)) * 100
    print(f"   {col}:")
    print(f"      Yes: {yes_count} ({yes_pct:.1f}%)")
    print(f"      No: {no_count} ({100-yes_pct:.1f}%)")

# Co-occurrence analysis
print(f"\n\n🔗 Co-occurrence of Conditions:")
df['Mental_Health_Issues'] = 0
for col in mental_health_cols:
    df['Mental_Health_Issues'] += (df[col] == 'Yes').astype(int)

condition_counts = df['Mental_Health_Issues'].value_counts().sort_index()
print("\n   Number of Conditions per Student:")
for i, count in enumerate(condition_counts.items()):
    num_conditions, num_students = count
    percentage = (num_students / len(df)) * 100
    print(f"   {int(num_conditions)} conditions: {num_students} students ({percentage:.1f}%)")

# Visualizations
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Mental Health Conditions Analysis', fontsize=16, fontweight='bold')

# Individual conditions
for idx, col in enumerate(mental_health_cols[:3]):
    ax = axes[idx//2, idx%2]
    data = df[col].value_counts()
    colors = ['#2ecc71', '#e74c3c']
    data.plot(kind='pie', ax=ax, autopct='%1.1f%%', colors=colors, startangle=90)
    ax.set_title(col, fontweight='bold', fontsize=11)
    ax.set_ylabel('')

# Co-occurrence bar chart
ax = axes[1, 1]
condition_counts.plot(kind='bar', ax=ax, color='steelblue', edgecolor='black')
ax.set_title('Co-occurrence of Mental Health Issues', fontweight='bold', fontsize=11)
ax.set_xlabel('Number of Conditions')
ax.set_ylabel('Number of Students')
ax.tick_params(axis='x', rotation=0)
ax.grid(alpha=0.3, axis='y')

plt.tight_layout()
plt.savefig('EDA_Mental_Health.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n✅ Mental health analysis plot saved as 'EDA_Mental_Health.png'")
```

## 1️⃣3️⃣ Gender vs Mental Health

```python
print("\n" + "="*70)
print("👫 GENDER vs MENTAL HEALTH ANALYSIS".center(70))
print("="*70)

print(f"\n📊 Gender & Mental Health Crosstab:")
for col in mental_health_cols:
    print(f"\n{col}:")
    crosstab = pd.crosstab(df['Choose your gender'], df[col], margins=True)
    print(crosstab)

# Visualization
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
fig.suptitle('Gender vs Mental Health Conditions', fontsize=16, fontweight='bold')

for idx, col in enumerate(mental_health_cols):
    ax = axes[idx]
    crosstab = pd.crosstab(df['Choose your gender'], df[col])
    crosstab.plot(kind='bar', ax=ax, color=['#2ecc71', '#e74c3c'], edgecolor='black')
    ax.set_title(col, fontweight='bold', fontsize=11)
    ax.set_xlabel('Gender')
    ax.set_ylabel('Count')
    ax.legend(title='Response', loc='upper right')
    ax.tick_params(axis='x', rotation=45)
    ax.grid(alpha=0.3, axis='y')

plt.tight_layout()
plt.savefig('EDA_Gender_MentalHealth.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n✅ Gender analysis plot saved as 'EDA_Gender_MentalHealth.png'")
```

## 1️⃣4️⃣ Course vs CGPA Analysis

```python
print("\n" + "="*70)
print("📚 COURSE vs CGPA ANALYSIS".center(70))
print("="*70)

print(f"\n📊 Courses in Dataset:")
course_counts = df['What is your course?'].value_counts()
for course, count in course_counts.items():
    percentage = (count / len(df)) * 100
    print(f"   {course}: {count} ({percentage:.1f}%)")

print(f"\n\n📊 CGPA Distribution by Course:")
# Convert CGPA string ranges to numeric (extract lower bound value)
df_cgpa = df.copy()
df_cgpa['CGPA_numeric'] = df_cgpa['What is your CGPA?'].str.split('-').str[0].astype(float)

cgpa_by_course = df_cgpa.groupby('What is your course?')['CGPA_numeric'].agg([
    ('Mean', 'mean'),
    ('Median', 'median'),
    ('Std', 'std'),
    ('Min', 'min'),
    ('Max', 'max')
]).round(2)
print(cgpa_by_course)

# Visualization
fig, axes = plt.subplots(1, 2, figsize=(16, 5))
fig.suptitle('Course vs CGPA Analysis', fontsize=16, fontweight='bold')

# Course distribution
ax = axes[0]
course_counts.plot(kind='barh', ax=ax, color='skyblue', edgecolor='black')
ax.set_title('Student Distribution by Course', fontweight='bold')
ax.set_xlabel('Number of Students')
ax.grid(alpha=0.3, axis='x')

# CGPA by course box plot (using numeric conversion)
ax = axes[1]
cgpa_data = [df_cgpa[df_cgpa['What is your course?'] == course]['CGPA_numeric'].dropna().values 
             for course in df_cgpa['What is your course?'].unique()]
bp = ax.boxplot(cgpa_data, labels=df_cgpa['What is your course?'].unique(), patch_artist=True)
for patch in bp['boxes']:
    patch.set_facecolor('lightblue')
ax.set_title('CGPA Distribution by Course', fontweight='bold')
ax.set_xlabel('Course')
ax.set_ylabel('CGPA')
ax.tick_params(axis='x', rotation=45)
ax.grid(alpha=0.3, axis='y')

plt.tight_layout()
plt.savefig('EDA_Course_CGPA.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n✅ Course-CGPA analysis plot saved as 'EDA_Course_CGPA.png'")
```

## 1️⃣6️⃣ Data Quality Assessment

```python
print("\n" + "="*70)
print("✅ DATA QUALITY ASSESSMENT".center(70))
print("="*70)

# Duplicate rows
duplicates = df.duplicated().sum()
print(f"\n📌 Duplicate Rows: {duplicates}")

# Data types
print(f"\n📌 Data Type Distribution:")
print(f"   • Numeric: {len(df.select_dtypes(include=[np.number]).columns)}")
print(f"   • Categorical: {len(df.select_dtypes(include=['object']).columns)}")

# Value consistency
print(f"\n📌 Categorical Values Consistency:")
for col in df.select_dtypes(include=['object']).columns:
    unique_vals = df[col].nunique()
    print(f"   {col}: {unique_vals} unique values")

# Data completeness
completeness = (1 - df.isnull().sum().sum() / (len(df) * len(df.columns))) * 100
print(f"\n📌 Overall Data Completeness: {completeness:.1f}%")

# Outliers
print(f"\n📌 Outlier Detection (IQR Method):")
for col in df.select_dtypes(include=[np.number]).columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = len(df[(df[col] < Q1 - 1.5*IQR) | (df[col] > Q3 + 1.5*IQR)])
    print(f"   {col}: {outliers} outliers")

print(f"\n" + "="*70)
print("✅ DATA QUALITY ASSESSMENT COMPLETE!".center(70))
print("="*70)
```

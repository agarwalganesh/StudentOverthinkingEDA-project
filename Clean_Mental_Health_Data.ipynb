# Student Mental Health Dataset - Data Cleaning

Clean the dataset by removing unnecessary columns using NumPy and Pandas

## 1️⃣ Import Required Libraries

```python
import pandas as pd
import numpy as np

print("✅ Libraries imported successfully!")
print(f"Pandas version: {pd.__version__}")
print(f"NumPy version: {np.__version__}")
```

## 2️⃣ Load the Dataset

```python
# Load the dataset
df = pd.read_csv('Student Mental health.csv')

print(f"📊 Dataset loaded successfully!")
print(f"Total rows: {len(df)}")
print(f"Total columns: {len(df.columns)}")
```

## 3️⃣ Explore the Dataset Structure

```python
# Display dataset shape and column information
print("📋 Dataset Shape:")
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}\n")

print("📝 Column Names:")
print(df.columns.tolist())

print("\n🔍 Dataset Info:")
print(df.info())

print("\n📊 First 5 rows:")
print(df.head())
```

## 4️⃣ Remove Unnecessary Columns

**Columns to Remove (❌):**
- Timestamp
- Marital status
- Did you seek any specialist for a treatment?

```python
# Define columns to remove
columns_to_remove = [
    'Timestamp',
    'Marital status',
    'Did you seek any specialist for a treatment?'
]

print(f"🗑️ Removing {len(columns_to_remove)} columns...")
print(f"Columns being removed: {columns_to_remove}\n")

# Remove the unnecessary columns using pandas drop()
df_cleaned = df.drop(columns=columns_to_remove)

print("✅ Columns removed successfully!")
print(f"Original shape: {df.shape}")
print(f"Cleaned shape: {df_cleaned.shape}")
```

## 5️⃣ Verify the Cleaned Dataset

```python
# Verify the cleaned dataset
print("✨ CLEANED DATASET VERIFICATION ✨\n")

print("📌 Final Columns:")
print(df_cleaned.columns.tolist())

print(f"\n📊 Dataset Shape: {df_cleaned.shape} (Rows, Columns)")

print("\n🔍 Dataset Info:")
print(df_cleaned.info())

print("\n📋 First 5 rows of cleaned dataset:")
print(df_cleaned.head())

print("\n✅ Data Cleaning Complete!")
print("\n📌 Columns Retained (for model making):")
print("   ✅ Depression / Anxiety / Panic attack (Target variables)")
print("   👍 Age, CGPA, Your current year of Study (Features)")
print("   🤏 Gender, Course (Optional features)")
```

## 🎁 Bonus: Save Cleaned Dataset

Optionally save the cleaned dataset for further analysis and model building

```python
# Save the cleaned dataset to a new CSV file
output_filename = 'Student_Mental_Health_Cleaned.csv'
df_cleaned.to_csv(output_filename, index=False)

print(f"💾 Cleaned dataset saved as: {output_filename}")
print(f"📊 Ready for model building with {len(df_cleaned)} records and {len(df_cleaned.columns)} features!")

# Display summary statistics
print("\n📈 Summary Statistics of Cleaned Data:")
print(df_cleaned.describe())
```
